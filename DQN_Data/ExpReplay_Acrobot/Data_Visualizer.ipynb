```python
sibling_lives11,sibling_scores11,time_11 = hkl.load('Data_Acrobot_E1_NN1')
sibling_lives12,sibling_scores12,time_12 = hkl.load('Data_Acrobot_E1_NN2')
sibling_lives21,sibling_scores21,time_21 = hkl.load('Data_Acrobot_E2_NN1')
#sibling_lives22,sibling_scores22,time_22 = hkl.load('Data_Acrobot_E2_NN2')
sibling_lives31,sibling_scores31,time_31 = hkl.load('Data_Acrobot_E3_NN1')
sibling_lives32,sibling_scores32,time_32 = hkl.load('Data_Acrobot_E3_NN2')
sibling_lives41,sibling_scores41,time_41 = hkl.load('Data_Acrobot_E4_NN1')
#sibling_lives42,sibling_scores42,time_42 = hkl.load('Data_Acrobot_E4_NN2')
sibling_lives51,sibling_scores51,time_51 = hkl.load('Data_Acrobot_E5_NN1')
#sibling_lives52,sibling_scores52,time_52 = hkl.load('Data_Acrobot_E5_NN2')
```

```python


# set width of bar
barWidth = 0.25
fig = plt.subplots(figsize =(12, 8))
 
# set height of bar
#episodes_mean = [np.mean(sibling_lives1),np.mean(sibling_lives2),'''np.mean(sibling_lives3)''',np.mean(sibling_lives4),np.mean(sibling_lives5),np.mean(sibling_lives6)''',np.mean(sibling_lives7)''',np.mean(sibling_lives8)]
episodes_mean_NN1 = [np.mean(sibling_lives11),np.mean(sibling_lives21),np.mean(sibling_lives31),np.mean(sibling_lives41),np.mean(sibling_lives51)]
#episodes_std = [np.std(sibling_lives1),np.std(sibling_lives2),'''np.std(sibling_lives3)''',np.std(sibling_lives4),np.std(sibling_lives5),np.std(sibling_lives5),np.std(sibling_lives6),np.std(sibling_lives7),np.std(sibling_lives8)]
episodes_std_NN1 = [np.std(sibling_lives11),np.std(sibling_lives21),np.std(sibling_lives31),np.std(sibling_lives41),np.std(sibling_lives51)]

# Set position of bar on X axis
br1 = np.arange(5)
br2 = [x + barWidth for x in br1]
 
# Make the plot
plt.bar(br1, episodes_mean_NN1, color ='r', width = barWidth,
        edgecolor ='grey', label ='Mean')
plt.bar(br2, episodes_std_NN1, color ='g', width = barWidth,
        edgecolor ='grey', label ='Standard Deviation')
 
# Adding Xticks
plt.xlabel('Variations', fontweight ='bold', fontsize = 10)
plt.ylabel('Episode Information', fontweight ='bold', fontsize = 10)
plt.xticks([r + barWidth for r in range(5)],
        ['E1(10k,64)','E2(10k,32)','E3(5k,64)','E4(5k,32)','E5(1,1)'],fontsize = 15)
 
plt.legend()
plt.show()
```
